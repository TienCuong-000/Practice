```python

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import math
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score    
from sklearn.model_selection import train_test_split
```

```python
class My_LinearRegression:
    def __init__(self,lr=0.01,iter=1000):
        self.lr=lr
        self.iter=iter
    def fit(self,X,y):
        samples,features=X.shape    
        self.w=np.zeros(features)
        self.b=0
        
        for i in range(self.iter):
            y_pred=np.dot(X,self.w)+self.b
            e=y_pred-y
            
            dw=(1/samples)*np.dot(X.T,e)
            db=(1/samples)*np.sum(e)
            
            self.w-=self.lr*dw
            self.b-=self.lr*db
    
    def predict(self,x_test):
        pred=np.dot(x_test,self.w)+self.b
        return pred
    def r2_score(self,y_test,y_pred):
        sse=(y_test-y_pred)**2
        mean=np.mean(y_test)
        sst=(y_test-mean)**2
        r2=1-(np.sum(sse))/(np.sum(sst))
        return r2
    def mse(self,y_test,y_pred):
        n=y_test.size
        return (1/n)*np.sum((y_test-y_pred)**2)
    
```

```python
diabetes=datasets.load_diabetes()
data=pd.DataFrame(data=diabetes.data,columns=diabetes.feature_names)
data['target']=diabetes.target
print(data.head())
```

```python
import math
scaler=StandardScaler()
x=data.drop('target',axis=1)
y=data['target']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=My_LinearRegression()
model.__init__(0.01,50000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
r2=model.r2_score(y_test,y_pred)
mse=model.mse(y_test,y_pred)
print('R2_Score : 'f"{r2*100:0.2f}","%")
print('RMSE : 'f"{math.sqrt(mse):0.2f}")
print(model.w)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
print('R2_Score : 'f"{r2*100:0.2f}","%")
print('RMSE : 'f"{math.sqrt(mse):0.2f}")
print(model.coef_)
```
